### Preprocessign Data
Want to clean and standardize text
- lowercasing
- remove punctuations
- tokenization
- stop word removal
- lemmatization

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# flagging numbers. Numbers can cause noise in topic modeling so we can turn on or off
def preprocess_text(text, remove_numbers=True):
    text = text.lower()
    if remove_numbers:
        text = re.sub(r'\d+', '', text)
        
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    tokens = text.split()
    
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    clean_text = ' '.join(tokens)
    
    return clean_text
```

```python
combined_df['clean_text'] = combined_df['text'].apply(lambda x: preprocess_text(x, remove_numbers=True))
```

### Topic Modeling Try Out

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
vectorizer = CountVectorizer(max_df=0.9, min_df=10, stop_words='english')
dtm = vectorizer.fit_transform(combined_df['clean_text'])
```

```python
n_topics = 10

lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(dtm)
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(lda, vectorizer.get_feature_names_out(), no_top_words)
```

Trying topic modeling by region!

```python
n_topics = 5
min_reviews = 50

region_topics = {}

for postal in combined_df['postal_code'].unique():
    region_data = combined_df[combined_df['postal_code'] == postal]
    
    if len(region_data) < min_reviews:
        continue
    
    vectorizer = CountVectorizer(max_df=0.9, min_df=10, stop_words='english')
    dtm = vectorizer.fit_transform(region_data['clean_text'])
    
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda.fit(dtm)
    
    region_topics[postal] = {
        'model': lda,
        'vectorizer': vectorizer,
        'data': region_data
    }
    
    print(f"Postal Code: {postal}")
    for topic_idx, topic in enumerate(lda.components_):
        top_words = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-11:-1]]
        print(f"  Topic {topic_idx}: {' '.join(top_words)}")
    print("\n")
```
